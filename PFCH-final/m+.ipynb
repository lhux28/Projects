```python
import json

id_list = []

with open('/Users/liwen/Documents/Github/Projects/PFCH-final/m+_chairs.json', 'r') as file:
    chair_file = json.load(file)
    # print(len(chair_file))
    # print(json.dumps(chair_file, indent=2))

    for item in chair_file:
        id = item['id']
        # print(id)
        id_list.append(id)

print(*sorted(id_list), sep="\n")
```

```python
import re 

chair_list = []

for each in chair_file:

    constituents = each["constituents"]
    # print(constituents)

    artist_list = []

    for entry in constituents:
        # print(entry['name'])
        individual_artist = entry['name']
        # artist_dictionary = {
        #     "artist": individual_artist
        # }
        # # print(artist_dictionary)
        # artist_list.append(individual_artist)

        # # print(artist_list)
        
        # artist = each.get('artist')
        title = each.get('title')
        date = each.get('beginDate')
        medium = each.get('medium')
    
        chair_entry = {
            "artist": individual_artist,
            "date": date,
            "medium": medium,
            "title": title,
            "museum": "m+"
        }

    chair_list.append(chair_entry)

print(len(chair_list))
print(json.dumps(chair_list, indent=4))
```
